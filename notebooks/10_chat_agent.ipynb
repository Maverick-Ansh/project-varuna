## Load the agent and chat

```python
from varuna.agent.llm import Agent
agent = Agent(four_bit=True)      # set False on A100/L4 (more VRAM)
```

```python
print(agent.chat("What's the flood outlook for tomorrow? Which wards are most urgent?"))
```

```python
print(agent.chat("What if 150 mm falls in 24 hours instead?"))
```

```python
print(agent.chat("Where should we dig to cut flooding the most with a 150,000 cubic metre budget?"))
```

```python
print(agent.chat("How accurate is this model, and where should I be cautious?"))
```
